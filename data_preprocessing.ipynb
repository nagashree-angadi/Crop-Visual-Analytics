```python
crops = pd.read_csv("./static/Data/Production_Crops_E_All_Data.csv", encoding = "ISO-8859-1")
crops_noflag = pd.read_csv("./static/Data/Production_Crops_E_All_Data_NOFLAG.csv", encoding = "ISO-8859-1")
crops_eflag = pd.read_csv("./static/Data/Production_Crops_E_Flags.csv", encoding = "ISO-8859-1")
regions = pd.read_csv("./static/Data/countries&regions.csv", encoding = "ISO-8859-1")
```

```python
# Dropping unnessasary columns
crops_noflag.drop(columns=["Element Code", "Unit"], inplace=True)
crops_noflag.rename(columns={"Item": "Crop", 
                             "Item Code": "Crop Code", 
                             "Area" : "Country",
                             "Area Code" : "Country Code",
                             "Element" : "Type"}, inplace = True)
crops_noflag.head()
```

```python
# Checking the null values in data set
crops_noflag.isnull().sum().reset_index()
```

```python
# Imputing the null values
crops_noflag.fillna(0, inplace=True)
```

```python
# Tranforming data in usable format
years = ['Y1961','Y1962', 'Y1963', 'Y1964', 'Y1965', 'Y1966', 'Y1967', 'Y1968', 'Y1969',
       'Y1970', 'Y1971', 'Y1972', 'Y1973', 'Y1974', 'Y1975', 'Y1976', 'Y1977',
       'Y1978', 'Y1979', 'Y1980', 'Y1981', 'Y1982', 'Y1983', 'Y1984', 'Y1985',
       'Y1986', 'Y1987', 'Y1988', 'Y1989', 'Y1990', 'Y1991', 'Y1992', 'Y1993',
       'Y1994', 'Y1995', 'Y1996', 'Y1997', 'Y1998', 'Y1999', 'Y2000', 'Y2001',
       'Y2002', 'Y2003', 'Y2004', 'Y2005', 'Y2006', 'Y2007', 'Y2008', 'Y2009',
       'Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014', 'Y2015', 'Y2016', 'Y2017','Y2018']

data = pd.melt(crops_noflag, id_vars=['Country Code', 'Country', 'Crop Code', 'Crop', 'Type'], value_vars=years)
data = data.rename( columns = {"variable":"Year", "value":"Value" })
data["Year"] = data["Year"].apply( lambda x: int(x[1:]))
```

```python
# Continent wise data
continent = ['Africa', 'Asia', 'Australia and New Zealand', 'Oceania', 
             'Northern America', 'Europe', 'South America']
continent_data = data[data["Country"].isin(continent)]
continent_data = continent_data.rename(columns={'Country Code': 'Continent Code', 'Country':'Continent'})

# Dropping aggregated values from countries column
agg = regions["Country Group"].unique()
country_data = data[~data["Country"].isin(agg)]

# Getting ISO code for country
country_data = country_data.merge(regions[["Country Code", "M49 Code"]],  
                                  how='inner', left_on="Country Code", 
                                  right_on="Country Code" )
```

```python
# Writing the data to local dir
country_data.to_csv("./static/Data/PreprocessedData.csv", index=False) 
continent_data.to_csv("./static/Data/PreprocessedDataContinent.csv", index=False) 
```
